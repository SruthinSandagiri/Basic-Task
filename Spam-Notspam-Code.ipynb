```python
import nltk
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
import re
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
y = data1["v1"]
```

```python
y
```

```python
y.value_counts()
```

```python
le = LabelEncoder()
y_enc = le.fit_transform(y)
```

```python
y_enc
```

```python
raw_text = data1["v2"]
```

```python
stop_words = nltk.corpus.stopwords.words("english")
for i in range(len(raw_text)):
    words = nltk.word_tokenize(raw_text[i])
    words = [w for w in words if w not in stop_words]
    raw_text[i] = " ".join(words)
```

```python
Lemmatizer = nltk.stem.WordNetLemmatizer()
for i in range(len(raw_text)):
    words = nltk.word_tokenize(raw_text[i])
    words = [Lemmatizer.lemmatize(w) for w in words]
    raw_text[i] = " ".join(words)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))
X_ngrams = vectorizer.fit_transform(raw_text)
```

```python
X_ngrams.shape
```

```python
X_train, X_test,y_train,y_test = train_test_split(X_ngrams,y_enc,test_size=0.2,random_state=42, stratify=y_enc)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
metrics.f1_score(y_test,y_pred)
```

```python
cm = metrics.confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier()
classifier2.fit(X_train,y_train)
y_pred2=classifier2.predict(X_test)
metrics.f1_score(y_test,y_pred2)
```

```python
cm2 = metrics.confusion_matrix(y_test,y_pred2)
```
